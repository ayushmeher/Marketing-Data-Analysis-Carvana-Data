```python
%%sql

-- Get the spend per category
select ac.channel_id, ac.category, AVG(spend.spend) as spend_per_categgory
FROM 
ad_channels_categorized ac 
JOIN spend ON spend.channel_id = ac.channel_id
GROUP BY ac.category
ORDER BY spend_per_categgory
;
```

```python
%%sql

-- Get the channel_id, category and avg_cost_per_click 
SELECT ac.channel_id, ac.category, AVG(spend.spend / c.click_count) AS avg_cost_per_click
FROM (
    SELECT channel_id, COUNT(*) AS click_count
    FROM first_clicks
    GROUP BY channel_id
) c
JOIN first_clicks cd ON c.channel_id = cd.channel_id
JOIN ad_channels_categorized ac ON cd.channel_id = ac.channel_id
JOIN spend ON spend.channel_id = cd.channel_id
GROUP BY ac.category
ORDER BY avg_cost_per_click
;
# if the customer  don't know about the product and they get to know about the product through social media. 
```

```python
%%sql
WITH monthly_sales AS (
  SELECT 
    campaign,
    strftime('%Y-%m', sale_datetime) as month,
    SUM(gross_profit) as total_profit
  FROM 
    ad_channels_categorized
  JOIN 
    clicks ON ad_channels_categorized.channel_id = clicks.channel_id
  JOIN 
    sales ON clicks.user_id = sales.user_id
  GROUP BY 
    campaign, month
  HAVING 
    month BETWEEN '2022-10' AND '2022-12'
),
mom_growth AS (
  SELECT 
    campaign,
    month,
    total_profit,
    ((total_profit - LAG(total_profit) OVER (PARTITION BY campaign ORDER BY month)) / LAG(total_profit) OVER (PARTITION BY campaign ORDER BY month)) * 100 as growth_rate
  FROM 
    monthly_sales
)
SELECT 
  campaign,
  month,
  growth_rate
FROM 
  mom_growth
ORDER BY 
  growth_rate DESC;
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your data is stored in a pandas DataFrame
data = {
    'campaign': ['Debit Dharma', 'RealTruck', 'HeadLedger', 'KnickKnack', '25+', 'Sedan', 'ManualBarter', 'Giving Vine', 'Truck', 'Convertible',
                 'RealTruck', '25+', 'ManualBarter', 'Giving Vine', 'Convertible', 'Truck', 'KnickKnack', 'HeadLedger', 'Debit Dharma', 'Sedan', 'Teen'],
    'month': ['2022-12']*10 + ['2022-11']*11,
    'growth_rate': [302.23, 293.02, 252.88, 242.09, 201.69, 198.21, 121.99, 115.88, 64.80, 61.99,
                    -55.62, -58.85, -59.28, -62.24, -66.42, -69.38, -75.47, -75.49, -81.45, -82.62, -100.67]
}

df = pd.DataFrame(data)

# Sort the DataFrame by month in ascending order
df['month'] = pd.to_datetime(df['month'])
df = df.sort_values(by='month', ascending=True)

# Plotting the line graph
plt.figure(figsize=(10, 6))
for campaign in df['campaign'].unique():
    campaign_data = df[df['campaign'] == campaign]
    plt.plot(campaign_data['month'], campaign_data['growth_rate'], label=campaign)

# Manually set x-axis ticks
plt.xticks(df['month'].unique(), [month.strftime('%Y-%m') for month in df['month'].unique()])

plt.xlabel('Month')
plt.ylabel('Growth Rate')
plt.title('Campaign Growth Rate Over Months')
plt.legend()
plt.grid(True)
plt.show()
```

```python
%%sql
WITH total_spend AS (
  SELECT 
    partner, 
    SUM(spend) as spend
  FROM 
    ad_channels_categorized 
  JOIN 
    spend ON spend.channel_id = ad_channels_categorized.channel_id 
  GROUP BY 
    partner
),
total_sales AS (
  SELECT 
    partner, 
    SUM(gross_profit) as sales
  FROM 
    ad_channels_categorized
  JOIN 
    clicks ON ad_channels_categorized.channel_id = clicks.channel_id
  JOIN 
    sales ON clicks.user_id = sales.user_id
  GROUP BY 
    partner
)
SELECT 
  total_spend.partner, spend, sales, 
  spend / sales as CAC
FROM 
  total_spend
JOIN 
  total_sales ON total_spend.partner = total_sales.partner
ORDER BY 
  CAC;
```

# Which category of channel has the best ROI? The worst? (ROI = gross profit divided by advertising spend)


```python
%%sql
WITH total_spend AS (
  SELECT 
    category, 
    SUM(spend) as spend
  FROM 
    ad_channels_categorized 
  JOIN 
    spend ON spend.channel_id = ad_channels_categorized.channel_id 
  GROUP BY 
    category
),
total_sales AS (
  SELECT 
    category, 
    SUM(gross_profit) as sales
  FROM 
    ad_channels_categorized
  JOIN 
    clicks ON ad_channels_categorized.channel_id = clicks.channel_id
  JOIN 
    sales ON clicks.user_id = sales.user_id
  GROUP BY 
    category
)
SELECT 
  total_spend.category, spend, sales, 
  sales/spend as ROI
FROM 
  total_spend
JOIN 
  total_sales ON total_spend.category = total_sales.category
ORDER BY 
  ROI
```

# On average, how many different channels do customers interact with prior to locking a car?


```python
%%sql
WITH user_channels AS (
  SELECT 
      c.user_id,
      COUNT(DISTINCT c.channel_id) as num_channels
  FROM 
      clicks c
  JOIN 
      locks_cleaned l
  ON 
      c.user_id = l.user_id
  WHERE 
      c.click_datetime < l.lock_datetime_mod
  GROUP BY 
      c.user_id)
SELECT 
    AVG(num_channels) 
FROM 
    user_channels;
```

# - What percentage of sales don't have a click within 90 days prior to the conversion? How can you explain this?


```python
%sql select * from sales;
```

```python
%sql select * from clicks;
```

```python
%%sql
 SELECT 
        s.sale_id,
        CASE 
            WHEN c.click_datetime IS NULL OR 
                 date(s.sale_datetime) - date(c.click_datetime) > 90 THEN 1
            ELSE 0
        END as no_click
    FROM 
        sales s
    LEFT JOIN 
        clicks c
    ON 
        s.user_id = c.user_id
    AND 
        c.click_datetime <= s.sale_datetime
    AND 
        date(s.sale_datetime) - date(c.click_datetime) <= 90
    order by no_click desc
```

```python
%%sql
WITH sales_clicks AS (
    SELECT 
        s.sale_id,
        CASE 
            WHEN c.click_datetime IS NULL OR 
                 date(s.sale_datetime) - date(c.click_datetime) > 90 THEN 1
            ELSE 0
        END as no_click
    FROM 
        sales s
    LEFT JOIN 
        clicks c
    ON 
        s.user_id = c.user_id
    AND 
        c.click_datetime <= s.sale_datetime
    AND 
        date(s.sale_datetime) - date(c.click_datetime) <= 90
)

SELECT 
    AVG(no_click) * 100 as percentage_no_click
FROM 
    sales_clicks;
```
